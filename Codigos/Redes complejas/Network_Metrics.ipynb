```python
import networkx as nx
import matplotlib.pyplot as plt
from networkx.algorithms import approximation
%config InlineBackend.figure_format = 'svg' # figuras alta calidad jupyter


## Erdos-renyi model
n = 100
p = 0.5
G = nx.erdos_renyi_graph(n, p, seed=10)

fig = plt.figure("Degree of a random graph", figsize=(6, 6))
# Create a gridspec for adding subplots of different sizes
axgrid = fig.add_gridspec(5, 4)

ax0 = fig.add_subplot(axgrid[0:3, :])
Gcc = G.subgraph(sorted(nx.connected_components(G), key=len, reverse=True)[0])
pos = nx.spring_layout(Gcc, seed=10396953)
nx.draw_networkx_nodes(Gcc, pos, ax=ax0, node_size=20)
nx.draw_networkx_edges(Gcc, pos, ax=ax0, alpha=0.4)
ax0.set_title("Random Network")
ax0.set_axis_off()
      
clustering_coefficient = approximation.average_clustering(G, trials=1000, seed=10)
average_shortest_path = nx.average_shortest_path_length(G)

print("Average Clustering Coefficient:", clustering_coefficient)
print("Average Shortest Path Length:", average_shortest_path)
```

```python
## Barabasi-Albert
n = 400
m = 10
G = nx.barabasi_albert_graph(n, m, seed=None, initial_graph=None)

fig = plt.figure("Degree of a random graph", figsize=(6, 6))
# Create a gridspec for adding subplots of different sizes
axgrid = fig.add_gridspec(5, 4)

ax0 = fig.add_subplot(axgrid[0:3, :])
Gcc = G.subgraph(sorted(nx.connected_components(G), key=len, reverse=True)[0])
pos = nx.spring_layout(Gcc, seed=10396953)
nx.draw_networkx_nodes(Gcc, pos, ax=ax0, node_size=20)
nx.draw_networkx_edges(Gcc, pos, ax=ax0, alpha=0.4)
ax0.set_title("Scale-free Network")
ax0.set_axis_off()
      
clustering_coefficient = approximation.average_clustering(G, trials=1000, seed=10)
average_shortest_path = nx.average_shortest_path_length(G)

print("Average Clustering Coefficient:", clustering_coefficient)
print("Average Shortest Path Length:", average_shortest_path)
```
