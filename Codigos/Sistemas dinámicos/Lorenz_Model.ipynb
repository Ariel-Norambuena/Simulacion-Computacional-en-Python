### Importamos las librerias

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import rcParams
from scipy.integrate import odeint

label_size = 16
mpl.rcParams['xtick.labelsize'] = label_size 
mpl.rcParams['ytick.labelsize'] = label_size 

rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = ['DejaVu Sans', 'Tahoma']

%config InlineBackend.figure_format = 'svg' # figuras alta calidad jupyter
```

### Implentación del metodo de Odeint de Python para resolver el modelo de Lorenz

```python
# Definir la función para la EDO del péndulo simple
def LorenzModel(X, t, sigma, rho, beta):
    x, y, z = X
    dxdt = [sigma*(y-x), x*(rho-z), x*y-beta*z]
    return dxdt

# Parámetros del modelo
sigma = 10
rho = 28
beta = 8/3

# Condiciones iniciales
x0 = 1    
y0 = 1        
z0 = 1
X0 = [x0, y0, z0]

# Vector de tiempo
t0 = 0
T = 40
Nt = 1000
t = np.linspace(t0, T, Nt)

# Resolvemos el modelo SIR utilizando odeint
sol = odeint(LorenzModel, X0, t, args=(sigma, rho, beta))
x = sol[:, 0] 
y = sol[:, 1]  
z = sol[:, 2]

# Graficamos las soluciones x(t), y(t), z(t)
#fig, ax = plt.subplots(figsize=(6, 4), tight_layout=True)
#ax.plot(x, y, z, linestyle='-',linewidth=2,color='blue',label='S')
#plt.xlabel(r"$x$", fontsize=16, color='black')
#plt.ylabel(r"$y$", fontsize=16, color='black')
#plt.zlabel(r"$z$", fontsize=16, color='black')
#plt.title('Modelo de Lorenz', fontsize=18)
#plt.show()

# Graficamos las soluciones s = S/N, i = I/N, r = R/N
fig, ax = plt.subplots(figsize=(6, 4), tight_layout=True)
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot(x, y, z)
plt.show()
```
